A. Buat 12 tuple, masing-masing berisi 4 data (int, float, string) dengan urutan acak

```python
data_tuples = (
    (42, "apel", 3.14, "buku"),
    ("meja", 7.5, 88, "kursi"),
    (15, 2.71, "laptop", 99),
    ("pensil", 100, 6.28, "tas"),
    (33, "kamera", 1.41, 77),
    (8.99, "sepatu", 50, 4.67),
    ("gitar", 21, 9.81, "drum"),
    (64, 0.5, "piano", "helm"),
    ("botol", 3.33, 12, "lampu"),
    (7, "papan", 5.55, 200),
    (11.11, "jam", 45, 2.22),
    ("piring", 30, 8.08, "sendok"),
)

print(f"Jumlah tuple: {len(data_tuples)}")
for i, t in enumerate(data_tuples):
    print(f"  Tuple {i+1}: {t}")
```

B. Konversi data tuple menjadi list

```python
data_list = [list(t) for t in data_tuples]

print("Hasil konversi ke list:")
for i, item in enumerate(data_list):
    print(f"  List {i+1}: {item}")
```

C. Tambahkan 2 data baru pada index sesuai digit terakhir NIM

```python
last_digit_nim = 2
insert_index = last_digit_nim if last_digit_nim != 0 else 10

new_data_1 = [25, "headset", 4.44, "mouse"]
new_data_2 = [37, 6.66, "monitor", "speaker"]

print(f"Digit terakhir NIM: {last_digit_nim}")
print(f"Index penambahan  : {insert_index}")
print(f"Data baru 1       : {new_data_1}")
print(f"Data baru 2       : {new_data_2}")

data_list.insert(insert_index, new_data_1)
data_list.insert(insert_index + 1, new_data_2)

print(f"\nList setelah penambahan (total: {len(data_list)}):")
for i, item in enumerate(data_list):
    print(f"  List {i+1}: {item}")
```

D. Hapus data bertipe float dari list

```python
data_no_float = []
for sublist in data_list:
    filtered = [item for item in sublist if not isinstance(item, float)]
    data_no_float.append(filtered)

print("List setelah float dihapus:")
for i, item in enumerate(data_no_float):
    print(f"  List {i+1}: {item}")
```

E. Urutkan elemen dalam list secara descending

```python
flat_list = []
for sublist in data_no_float:
    for item in sublist:
        flat_list.append(item)

int_items = sorted([x for x in flat_list if isinstance(x, int)], reverse=True)
str_items = sorted([x for x in flat_list if isinstance(x, str)], reverse=True)

sorted_list = int_items + str_items

print(f"Elemen integer (descending) : {int_items}")
print(f"Elemen string  (descending) : {str_items}")
print(f"\nList gabungan terurut descending:")
print(f"  {sorted_list}")
```

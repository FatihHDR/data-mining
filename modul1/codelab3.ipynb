```python
import pandas as pd
```

A. Series – Nama Produk & ID Produk

```python
produk_dict = {
    "Laptop Gaming": "PRD-001",
    "Mouse Wireless": "PRD-002",
    "Keyboard Mechanical": "PRD-003",
    "Monitor 24 inch": "PRD-004",
    "Headset Bluetooth": "PRD-005",
}

series_produk = pd.Series(produk_dict, name="ID Produk")
series_produk.index.name = "Nama Produk"

print(series_produk)
print(f"\nTipe data: {type(series_produk)}")
```

B. DataFrame – 4 Kolom × 5 Baris

```python
data_dict = {
    "Nama Produk": ["Laptop Gaming", "Mouse Wireless", "Keyboard Mech", "Monitor 24\"", "Headset BT"],
    "Harga (Rp)":  [15_000_000, 350_000, 850_000, 2_500_000, 450_000],
    "Stok":        [10, 50, 30, 20, 40],
    "Kategori":    ["Komputer", "Aksesori", "Aksesori", "Komputer", "Audio"],
}

df = pd.DataFrame(data_dict)
print("DataFrame awal:")
print(df.to_string(index=True))
```

C. Menambahkan 1 Kolom Baru

```python
df["Diskon (%)"] = [10, 5, 15, 20, 10]

print("DataFrame setelah ditambah kolom 'Diskon (%)':")
print(df.to_string(index=True))
```

D. Mengurutkan DataFrame

```python
# Sort berdasarkan Nama Produk (A → Z)
df_sorted_nama = df.sort_values(by="Nama Produk", ascending=True).reset_index(drop=True)
print("Diurutkan berdasarkan Nama Produk (A–Z):")
print(df_sorted_nama.to_string(index=True))

# Sort berdasarkan Harga (tertinggi → terendah)
df_sorted_harga = df.sort_values(by="Harga (Rp)", ascending=False).reset_index(drop=True)
print("\nDiurutkan berdasarkan Harga (tertinggi–terendah):")
print(df_sorted_harga.to_string(index=True))
```

```python
courses_tuple = ({
    "Matematika": 88,
    "Fisika": 76,
    "Kimia": "",
    "Biologi": 91,
    "Sejarah": "",
    "Bahasa Inggris": 84,
    "Ekonomi": 79,
    "Pemrograman": 95,
    "Statistika": "",
    "Data Mining": 89
},)
```

```python
courses_dict = courses_tuple[0]

cleaned_courses = {k: v for k, v in courses_dict.items() if v != "" and v is not None}

remaining_count = len(cleaned_courses)

sorted_courses = sorted(cleaned_courses.items(), key=lambda item: item[1])

highest_course = max(cleaned_courses.items(), key=lambda item: item[1])

average_grade = sum(cleaned_courses.values()) / remaining_count if remaining_count > 0 else 0
```

```python
print("Tuple berisi dictionary:")
print(courses_tuple)

print("\n(a) Data setelah menghapus nilai kosong:")
print(cleaned_courses)

print("\n(b) Jumlah entri tersisa:", remaining_count)

print("\n(c) Data terurut berdasarkan nilai (ascending):")
print(sorted_courses)

print("\n(d) Mata kuliah dengan nilai tertinggi:")
print(f"{highest_course[0]} = {highest_course[1]}")

print("\n(e) Rata-rata nilai:")
print(round(average_grade, 2))
```

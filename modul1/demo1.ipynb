```python
import random


def demo1_operation(data_list, nim):
    result = data_list.copy()

    for _ in range(10):
        if random.choice([True, False]):
            result.append(random.randint(-100, 100))
        else:
            result.append(round(random.uniform(-100, 100), 2))

    remove_odd_index = nim % 2 == 1
    filtered = [
        value
        for idx, value in enumerate(result)
        if not ((idx % 2 == 1) if remove_odd_index else (idx % 2 == 0))
    ]

    filtered.reverse()

    if filtered:
        filtered.remove(max(filtered))
    filtered.sort(reverse=True)

    return filtered


nim = 202310370311132
data_awal = [5, 12.5, -3, 20, 7.7]
hasil = demo1_operation(data_awal, nim)

print("Data awal:", data_awal)
print("Hasil akhir:", hasil)
```
